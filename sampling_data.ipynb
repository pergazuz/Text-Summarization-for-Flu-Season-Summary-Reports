```python
df = pd.read_csv('data/flu_data_summary_qa_length.csv')
df
```

```python
import pandas as pd


bins = [0, 250, 500, 750, 1000]

# Define the labels for the bins
labels = ['[0, 250)', '[250, 500)', '[500, 750)', '[750, 1000)']

df['Word_Count_Group'] = pd.cut(df['Word_Count'], bins=bins, labels=labels, right=False)

print(df[['Year', 'Topic', 'Word_Count', 'Word_Count_Group']])
```

```python
# Calculate the desired initial samples for each group based on available sizes
group_counts = df['Word_Count_Group'].value_counts()
print(group_counts)
```

```python
grouped = df.groupby('Word_Count_Group')

sample_size = 10
percent_to_sample = 0.2

samples_per_group = (grouped.size() * percent_to_sample).apply(np.ceil).astype(int)

while samples_per_group.sum() > sample_size:
    largest_group = samples_per_group.idxmax()
    samples_per_group[largest_group] -= 1

sampled_data = pd.concat([
    group.sample(n if n <= len(group) else len(group), random_state=1) 
    for (name, group), n in zip(grouped, samples_per_group)
])

sampled_data
```

```python
sampled_data.to_csv('data/flu_data_sampled.csv', index=False)
```
